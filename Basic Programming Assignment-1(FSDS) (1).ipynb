## 2. Write a Python program to do arithmetical operations addition and division.?

```python
a=int(input("Enter a's value : "))
b=int(input("Enter b's value : "))
print("Sum of ",a,"&",b,"is :",a+b)
print("Division of ",a,"&",b,"is :",a/b)
```

## 3. Write a Python program to find the area of a triangle?

```python
base=int(input("Enter base value : "))
height=int(input("Enter height value : "))
area=(1/2)*base*height
print("Area of triangle for the given base and height is :",area)
```

## 4. Write a Python program to swap two variables?

```python
a=int(input("Enter a's value : "))
b=int(input("Enter b's value : "))
c=a
a=b
b=c
print("Swapped values of a,b are :",a,b)
```
