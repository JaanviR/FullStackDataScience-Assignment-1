## 1. Write a Python program to find sum of elements in list?

```python
l=[1,2,5,88,100,200]
sum=0
for i in l:
    sum=sum+i
print("Sum of all elements inside the list is : ",sum)
```

## 5. Write a Python program to find second largest number in a list?


```python
l3=[35,88,100,9,1000,355456,355456,355456]
l3.sort()
l3
for i in range(len(l3)-1,-1,-1):
    if l3[i-1] == l3[i]:
        continue
    else:
        print("Second largest number in the given list ",l3, "is",l3[i-1])
        break
```

## 6. Write a Python program to find N largest elements from a list?

## 7. Write a Python program to print even numbers in a list?

```python
l=[88,95,3454,1,7567]
l2=[]
for i in l:
    if i%2==0:
        l2.append(i)
print("List of even numbers found in the given list",l, "are : ",l2)
```

## 9. Write a Python program to Remove empty List from List?

```python
l=[[],[1,2,3],"jaanvi",[],[],[],7,8,12,[]]
l5=[]
for i in l:
    if type(i)==list:
        if len(i)==0:
            l5.append(i)
print("List of empty lists ",l5)
for i in l:
    for j in l5:
        if i == j:
            l.remove(i)
print("Original list after removing empty lists :",l)
```

## 11. Write a Python program to Count occurrences of an element in a list

```python
l=[[1,2,3],66,"jaanvi",7,8,12,[1,"jaanvir,3"],[1,2,3],"yuvi",66,"jaanvi"]
for i in l:
    k=l.count(i)
    if k==1:
        print("Element",i,"is repeated --->",k,"time")
    else:
        print("Element",i,"is repeated --->",k,"times")
```
