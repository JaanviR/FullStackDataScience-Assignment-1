## 1. Write a Python program to convert kilometers to miles?

```python
n = int(input("Enter a value in 'km' to convert it into 'miles' : "))
q = n*0.612
print("Given value in miles is ",q)
```

## 2. Write a Python program to convert Celsius to Fahrenheit?

```python
n = int(input("Enter a value in 'Celsius' to convert it into 'Fahrenheit' : "))
k = (n*(9/5))+32
print("Given value in Fahrenheit is ",k)
```

## 4. Write a Python program to solve quadratic equation?

```python
import cmath
```

```python
a = int(input("Enter a's value : "))
b = int(input("Enter b's value : "))
c = int(input("Enter c's value : "))
e = str(a)+"x2"+"+"+str(b)+"x"+"+"+str(c)+"= 0"
print("The quadratic equation formed with given variables is : ",e)

k=(b**2)-(4*a*c)

print("k value is",k)

if k < 0:
    print("The given equation",e,"has no real solution")
elif k == 0:
    sol1 = (-b+cmath.sqrt(k))/(2*a)
    print("The given equation",e,"has only one solution",sol1)
else:
    sol1 = (-b+cmath.sqrt(k))/(2*a)
    sol2 = (-b-cmath.sqrt(k))/(2*a)
    print("Solutions of given quadratic equation",e,"are : ",sol1,sol2)
```

## 5. Write a Python program to swap two variables without temp variable?


```python
a = int(input("Enter a's value : "))
b = int(input("Enter a's value : "))
print("Values of a and b before swaping are : ","a =",a,"; b =",b)
a,b = b,a
print("Values of a and b after swaping are : ","a =",a,"; b =",b)
```
