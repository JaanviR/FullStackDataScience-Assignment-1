## 1. Write a Python Program to Find LCM?

```python
n1 = int(input("Enter n1 value: "))
n2 = int(input("Enter n2 value: "))
def lcm(a,b):
    if a > b:
        max = a
    else:
        max = b
    while True:
        if max%a == 0 and max%b == 0:
            lcm = max
            break
        max+=1
    return lcm
print("LCM of given two numbers is", lcm(n1,n2))
```

## 2. Write a Python Program to Find HCF?

```python
n1 = int(input("Enter n1 value: "))
n2 = int(input("Enter n2 value: "))
def hcf(a,b):
    l=[]
    for i in range(1, a+b+1):
        if a%i==0 and b%i==0:
            l.append(i)
    l.sort()
    return max(l)
print("HCF of given two numbers is", hcf(n1,n2))
```

## 3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?


```python
n = int(input("Enter a number : "))
print("Given value",n, "in binary format is :", bin(n).replace("0b",''))
print("Given value",n, "in octal format is :", oct(n).replace("0o",''))
print("Given value",n, "in hexadecimal format is :", hex(n).replace("0x",''))
```

## 4. Write a Python Program To Find ASCII value of a character?

```python
n = input("Enter a character : ")
print("ASCII value of given character is : ",ord(n))
```

## 5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
a=int(input("Enter a's value : "))
b=int(input("Enter b's value : "))
def sum(a,b):
    return a+b
def dif(a,b):
    return a-b
def mul(a,b):
    return a*b
def div(a,b):
    return a//b

print("Sum is :", sum(a,b))
print("Difference is :", dif(a,b))
print("Multiplication is :", mul(a,b))
print("Division is :", div(a,b))
```
