## 1. Write a Python Program to Find the Factorial of a Number?

```python
num = int(input("Enter a number : "))
prod = 1
for i in range(1,num+1):
    prod = prod * num
    num = num - 1
print("Factorial of a given number is :", prod)  
```

## 2. Write a Python Program to Display the multiplication Table?

```python
num = int(input("Enter a number for the multiplication table to be displayed : "))
n = int(input("Enter the range : "))
print("Multiplication table of given number ",num, "is")
for i in range(1,n+1):
    print(num,"*",i,"=",num*i)
```

## 3. Write a Python Program to Print the Fibonacci sequence?

```python
num = int(input("Enter a number : "))
a=1
a,b = 1,a
l=[a]
i = 0
while i < num:
    c = a + b
    a = b
    b = c
    l.append(b)
    i+=1
print(l)
```

## 4. Write a Python Program to Check Armstrong Number?

```python
# function to count number of digits 
def countdigits(a):
    a = str(a)
    count = 0
    for i in a:
        count = count+1
    return count
```

```python
num = int(input("Enter a number to check whether it is armstrong number or not : "))
q = num
cd = countdigits(q)
sum = 0
while q!=0:
    k = (q%10)
    q = int((q/10)) # or q//10 --> returns int value
    sum = sum+(k**cd)
if sum == num:
    print("Given number", num, "is Armstrong number")
else:
    print("Given number", num, "is not Armstrong number")
```

## 5. Write a Python Program to Find Armstrong Number in an Interval?

```python
def armstrong(num):
    q = num
    cd = countdigits(q)
    sum = 0
    while q!=0:
        k = (q%10)
        q = int((q/10))
        sum = sum+(k**cd)
    if sum == num:
        print(num)
        
lval = int(input("Enter lower bound value : "))
uval = int(input("Enter upper bound value : "))
for i in range(lval,uval+1):
    armstrong(i)
```
