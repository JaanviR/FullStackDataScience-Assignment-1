## 7. Describe a condition and when you would use one.
Conditional statements perform actions or execute statements based on the boolean constraints(true or false).
For eg: If we would like to give access to an account after checking whether the person is authenticated or not then we can write a condition as shown below


```python
Username = str(input("Enter registered username: "))
if Username == "jaanvi":
    print("You are authenticated and you can access the account")
else:
    print("Username is not registered, try again!!")
```

In the above example we are evaluating a condition(username=="jaanvi") by checking whether the name matches with the user input or not. If the name matches with the user input then the condition (username == "jaanvi") evaluates to "True" and execute the print statement inside the "if-block" if the condition evaluates to "False" then the control goes to the "else" block and execute the statements inside that else block.

## 8. Recognize the following three blocks in this code


```python
spam = 0 #declaring/initializing the variable
if spam == 10: #if block -->conditional statement
    print('eggs')#print statement for outer if block
    if spam > 5: #another if block(nested if)
        print('bacon')
    else: #else block for inner if block(gets execute only if the condition is flase)
        print('ham')
    print('spam')#print statement for outer if block
print('spam')#statement-->outside of if block
```

## 9. Create a programme that prints. If 1 is stored in spam, prints Hello; if 2 is stored in spam, prints Howdy; and if 3 is stored in spam, prints Salutations! if there's something else in spam.

```python
spam = int(input("Enter spam value : "))
if spam == 1:
    print("Hello")
elif spam == 2:
    print("Howdy")
elif spam == 3:
    print("Salutations!")
else:
    print("Spam!!")
```
